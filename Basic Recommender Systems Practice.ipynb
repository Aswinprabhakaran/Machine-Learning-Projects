```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
column_names = ['user_id', 'item_id', 'rating', 'timestamp']
df = pd.read_csv('u.data',sep='\t',names=column_names)
```

```python
df.head()
```

```python
movie_titles = pd.read_csv('Movie_Id_Titles')
movie_titles.head()
```

```python
final = pd.merge(df,movie_titles,on='item_id')
final.head()
```

```python
final.groupby('title')['rating'].mean()
```

```python
final.groupby('title')['rating'].count()
```

```python
ratings = pd.DataFrame(final.groupby('title')['rating'].mean())
ratings ['num of ppl who gave ratings'] = pd.DataFrame(final.groupby('title')['rating'].count())
ratings.head()
```

```python
ratings.sort_values('num of ppl who gave ratings',ascending=False)
```

```python
indu_ratings_to_movies = final.pivot_table(index='user_id',columns='title',values='rating')
indu_ratings_to_movies.head()
```

```python
star_wars=indu_ratings_to_movies['Star Wars (1977)']
liar_liar= indu_ratings_to_movies['Liar Liar (1997)']
```

```python
similar_to_starwars = indu_ratings_to_movies.corrwith(star_wars)
similar_to_liarliar = indu_ratings_to_movies.corrwith(liar_liar)
```

```python
corr_liarliar = pd.DataFrame(similar_to_liarliar,columns=['Correlation'])
corr_liarliar.dropna(inplace=True)
corr_liarliar = corr_liarliar.join(ratings['num of ppl who gave ratings'])
corr_liarliar[corr_liarliar['num of ppl who gave ratings']>100].sort_values('Correlation',ascending=False)
```
