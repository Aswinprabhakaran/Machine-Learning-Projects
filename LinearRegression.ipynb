```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df1=pd.read_csv('train.csv')
df2=pd.read_csv('features.csv')
```

```python
df1.head()
```

```python
df2.head()
```

```python
df2.head()
```

```python
df1=pd.merge(df1,df2,on=['Store','Date','IsHoliday'])
```

```python
df1['IsHoliday'] = pd.get_dummies(df1['IsHoliday'],drop_first=True)
df1.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df1[['Store','Dept','Year','Month','Day','Temperature','Fuel_Price','CPI','Unemployment','IsHoliday']]
Y=df1['Weekly_Sales']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.40)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm= LinearRegression()
```

```python
lm.fit(X_train,Y_train)
```

```python
pd.DataFrame(lm.coef_,index=X_train.columns)
```

```python
predictions = lm.predict(X_test)
```

```python
lm.score(X,Y)
```
